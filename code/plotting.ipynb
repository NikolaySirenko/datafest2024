```python
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../finalMerge.csv", delimiter=',')
df.info()
df.columns
```

```python
averages = df['Opening_duration_in_seconds'].groupby(df['Opening_duration_in_seconds'].index // 365).mean()

print(averages)
```

```python
# Plot the original series
plt.plot(averages)
plt.title('Opening_duration_in_seconds')
plt.xlabel('Hour')
plt.ylabel('Value')
plt.show()
```

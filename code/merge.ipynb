```python
import pandas as pd
import numpy as np
import os
```

```python
# Adjust pandas display options to ensure all columns are shown
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.width', None)
pd.set_option('display.max_rows', None)
```

```python
csv_files = []
# Walk the directory tree
for root, dirs, files in os.walk('../files_for_merge'):
    for file in files:
        # Check if the file ends with '.csv'
        if file.endswith('.csv'):
            # Get the full path to the file
            full_path = os.path.join(root, file)
            csv_files.append(full_path)
csv_files
```

```python
df_file_path = (csv_files[0])
df_bridgeOpening = pd.read_csv(df_file_path, delimiter=',')
column_mapping = {"Hour of the year":"Hour of year"}
df_bridgeOpening = df_bridgeOpening[['Opening_duration_in_seconds', 'Hour of the year']].rename(columns=column_mapping)
df_bridgeOpening.info()
df_bridgeOpening.head(10)
```

```python
df_file_path = (csv_files[1])
df_holidays = pd.read_csv(df_file_path, delimiter=',')
df_holidays.info()
df_holidays.head(10)
```

```python
df_file_path = (csv_files[2])
df_weather = pd.read_csv(df_file_path, delimiter=',')
df_weather.info()
df_weather.head(10)
```

```python
df_merged = pd.merge(df_weather, df_holidays, on='Date', how='left')
df_merged.info()
df_merged.head(10)
```

```python
df_merged = pd.merge(df_merged, df_bridgeOpening, on='Hour of year', how='left')
df_merged.info()
```

```python
# Replace all NaN values with 0
df_merged.fillna(0, inplace=True)
df_merged.head(10)
```

```python
# Export the DataFrame to a new CSV file
df_merged.to_csv('../mergedData\\Weather+Holidays+Openings.csv', index=False)
```

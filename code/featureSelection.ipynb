```python
df_file_path = (csv_files[4])
df = pd.read_csv(df_file_path, delimiter=',')
df.info()
```

```python
df['Country'].unique()
```

```python
dates = df['Date'].unique()
dates
```

```python
df_holidays = pd.DataFrame(columns=['Date', 'NL', 'BE', 'DE', 'FR'])
df_holidays
```

```python
countries = df['Country'].unique()
countries
```

```python
for date in dates:
    new_row = [date]
    for country in countries:
        if df[df['Date'] == date]['Country'].isin([country]).any():
            new_row.append(1)
        else:
            new_row.append(0)
    print(new_row)
    df_holidays.loc[len(df_holidays)] = new_row
df_holidays
```

```python
column_mapping = {
    'NL':'NL national holiday', 
    'BE':'BE national holiday', 
    'DE':'DE national holiday', 
    'FR':'FR national holiday'
}
df_holidays = df_holidays.rename(columns=column_mapping)
df_holidays
```

```python
df_file_path = (csv_files[5])
df = pd.read_csv(df_file_path, delimiter=',')
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
df.info()
```

```python
df['Country'].unique()
```

```python
dates = df['Date'].unique()
dates
```

```python
df_holidays2 = pd.DataFrame(columns=['Date', 'NL', 'BE', 'DE', 'FR'])
df_holidays2
```

```python
countries2 = ['NL', 'BE', 'DE', 'FR']
for date in dates:
    new_row = [date]
    for country in countries2:
        if df[df['Date'] == date]['Country'].isin([country]).any():
            new_row.append(1)
        else:
            new_row.append(0)
    print(new_row)
    df_holidays2.loc[len(df_holidays2)] = new_row
df_holidays2
```

```python
merged_df = pd.merge(df_holidays, df_holidays2, on='Date', how='outer')
```

```python
merged_df
```

```python
# Replace NaN values with 0
merged_df.fillna(0, inplace=True)
column_mapping = {
    'NL':'NL school holiday', 
    'BE':'BE school holiday', 
    'DE':'DE school holiday', 
    'FR':'FR school holiday'
}
merged_df = merged_df.rename(columns=column_mapping)
merged_df
```

```python
# Export the DataFrame to a new CSV file
merged_df.to_csv('../featureSelection\\holidaysPerDate.csv', index=False)
```

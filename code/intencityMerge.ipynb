```python
import pandas as pd
import numpy as np
import os
```

```python
df_environment = pd.read_csv('../mergedData/Weather+Holidays+Openings.csv', delimiter=',')
df_environment.info()
```

```python
csv_files = []
# Walk the directory tree
for root, dirs, files in os.walk('../intensityForMerge'):
    for file in files:
        # Check if the file ends with '.csv'
        if file.endswith('.csv'):
            # Get the full path to the file
            full_path = os.path.join(root, file)
            csv_files.append(full_path)
csv_files
```

```python
new_csv_files = [os.path.basename(file_path) for file_path in csv_files]
new_csv_files = [os.path.join('../intensityMerge', filename) for filename in new_csv_files]
new_csv_files
```

```python
df_file_path = (csv_files[2])
df = pd.read_csv(df_file_path, delimiter=',')
df.info()
```

```python
column_mapping = {'object intensity ': 'object intensity West 2023',"Hour of the year":"Hour of year"}
df = df[['object intensity ', 'Hour of the year']].rename(columns=column_mapping)
df.info()
```

```python
df_merged = pd.merge(df_environment, df, on='Hour of year', how='left')
df_merged.info()
df_merged.head(100)
```

```python
df_file_path = (csv_files[3])
df = pd.read_csv(df_file_path, delimiter=',')
df.info()
```

```python
column_mapping = {'object intensity ': 'object intensity East 2023',"Hour of the year":"Hour of year"}
df = df[['object intensity ', 'Hour of the year']].rename(columns=column_mapping)
df.info()
```

```python
df_merged = pd.merge(df_merged, df, on='Hour of year', how='left')
df_merged.info()
df_merged.head(100)
```

```python
df_merged.fillna(0, inplace=True)
```

```python
df_merged.info()
```

```python
# Export the DataFrame to a new CSV file
df_merged.to_csv('../finalMerge.csv', index=False)
```
